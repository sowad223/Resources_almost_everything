```python
class Calculator:
  def __init__(self):
    self.value1=None
    self.value2=None
    self.sign=''
  def calculate(self,value1,value2,sign):
    self.value1=value1
    self.value2=value2
    self.sign=sign
    if self.sign=='+':
      val=self.value1+self.value2
    elif sign=='-':
     val=self.value1-self.value2
    elif sign=='*':
     val=self.value1*self.value2
    elif self.sign=='/':
      val=self.value1/self.value2
    else:
      val="invalid"
    return val
  def showCalculation(self):
    print(f"{self.value1} {self.sign} {self.value2} = {val}")
c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
class Account:
  def __init__(self,*args):
    if len(args)==2:
      self.name=args[0]
      self.balance=float(args[1])

    elif len(args)==1:
      self.name=args[0]
      self.balance=0.0
    else:
      self.name="Default Account"
      self.balance=0.0
  def details(self):
    return f"{self.name}\n{self.balance}"
  def withdraw(self,amount):
    self.min_amount=3070
    self.money=self.balance-amount
    if self.money>self.min_amount:
      print(f"Withdraw successful! New balance is:{self.balance-amount}")
    else:
      print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")

a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```
