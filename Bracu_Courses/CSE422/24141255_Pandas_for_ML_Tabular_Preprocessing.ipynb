### Option 2: Using SimpleImputer to estimate a value for null values using existing values in the column

````python
sales.head()
````

````python
sales['salt'].isnull().sum()
````

````python
from sklearn.impute import SimpleImputer

impute = SimpleImputer(missing_values=np.nan, strategy='mean')

impute.fit(sales[['salt']])

sales['salt'] = impute.transform(sales[['salt']])  #Here we are making sure it is inplace = True
````

````python
sales['salt'].isnull().sum()
````

## Handling Categorical / Text Data

There are two types of text data we might come across.


*   Nominal - Not ordered. Non Heirchical. No ranking of one above another. For eg. names of countries, places, foods.
*   Ordinal - Ordered. Heirchical. Ranking of one above another. For eg. True / False, Best/Average/Worst



### Option 1: Nominal (No Ranking)


`
One hot encoding
`



````python
volunteer['category_desc'].unique()
````

````python
# Transform the category_desc column
category_enc = pd.get_dummies(volunteer['category_desc'])

# Take a look at the encoded columns
category_enc.head()
````

### Option 2: Ordinal (Ranking)

````python
hiking.head()
````



> **Using LabelEncoder**



````python
from sklearn.preprocessing import LabelEncoder

# Set up the LabelEncoder object
enc = LabelEncoder()

# Apply the encoding to the "Accessible" column
hiking['Accessible_enc'] = enc.fit_transform(hiking['Accessible'])   # Here we are not changing inplace

# Compare the two columns
print(hiking[['Accessible', 'Accessible_enc']].head())
````



> **Using map**



````python
hiking['Accessible_enc'] = hiking['Accessible'].map({'Y':1,'N':0})
````
